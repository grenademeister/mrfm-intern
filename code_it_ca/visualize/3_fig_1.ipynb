```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
from scipy.io import loadmat
```

```python
log_v = "log_downstream_recon"
run_idx = 11
run_ep = 48
log_root = Path(f"/home/juhyung/data/fm2026/log/log_downstream_recon/{run_idx:05d}_train/test/ep_{run_ep}")
res_files = sorted(log_root.glob("*.mat"))
print(len(res_files))
```

```python
idx = 0
```

```python
idx += 1
print(idx)

idx = 213

res_mat = loadmat(res_files[idx])
input = res_mat["input"].squeeze()
out = res_mat["out"].squeeze()
label = res_mat["label"].squeeze()

# x1, x2, y1, y2 = 70, -50, 120, -120
x1, x2, y1, y2 = 1, -1, 1, -1

vmax_out = np.percentile(out, 98) * 1.2
vmax_out = 4.0
vmin = 0

print(out.min(), out.max())

fig, ax = plt.subplots(figsize=(3, 3))
fig.patch.set_facecolor("black")
plt.imshow(input[x1:x2, y1:y2], cmap="gray", vmin=0, vmax=vmax_out)
plt.axis("off")
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(3, 3))
fig.patch.set_facecolor("black")
plt.imshow(out[x1:x2, y1:y2], cmap="gray", vmin=0, vmax=vmax_out)
plt.axis("off")
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(3, 3))
fig.patch.set_facecolor("black")
plt.imshow(label[x1:x2, y1:y2], cmap="gray", vmin=0, vmax=vmax_out)
plt.axis("off")
plt.tight_layout()
plt.show()


fig, ax = plt.subplots(figsize=(3, 3))
fig.patch.set_facecolor("black")
plt.imshow(abs(out[x1:x2, y1:y2] - label[x1:x2, y1:y2]), cmap="gray", vmin=0, vmax=vmax_out / 2)
plt.axis("off")
plt.tight_layout()
plt.show()
```
