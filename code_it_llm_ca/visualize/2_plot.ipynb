```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
psnr_fm_mean = [30.95, 32.52, 33.76, 35.18]
psnr_fm_std = [1.98, 2.28, 2.49, 2.67]
ssim_fm_mean = [0.7502, 0.8395, 0.8803, 0.9056]
ssim_fm_std = [0.0399, 0.0295, 0.0253, 0.0227]

psnr_sc_mean = [27.95, 31.33, 33.66, 35.43]
psnr_sc_std = [1.89, 2.15, 2.45, 2.70]
ssim_sc_mean = [0.5619, 0.8051, 0.8646, 0.9048]
ssim_sc_std = [0.0784, 0.0327, 0.0276, 0.0231]

psnr_unet_mean = [28.21, 30.56, 33.53, 35.36]
psnr_unet_std = [1.88, 2.02, 2.43, 2.69]
ssim_unet_mean = [0.6593, 0.7824, 0.8571, 0.9001]
ssim_unet_std = [0.0430, 0.0412, 0.0274, 0.0248]
```

```python
x = np.array([3, 10, 31, 100])

plt.figure(figsize=(10, 4.5))

plt.subplot(1, 2, 1)
plt.errorbar(x, psnr_sc_mean, yerr=psnr_sc_std, fmt="-o", color="tab:blue", markersize=6, capsize=4, label="From scratch")
plt.errorbar(x, psnr_unet_mean, yerr=psnr_unet_std, fmt="-s", color="tab:green", markersize=6, capsize=4, label="UNet")
plt.errorbar(x, psnr_fm_mean, yerr=psnr_fm_std, fmt="-^", color="tab:orange", markersize=6, capsize=4, label="FM")
plt.xscale("log")
plt.xlabel("Training Volumes (log scale)")
plt.ylabel("PSNR")
plt.title("PSNR")
plt.legend()
plt.grid(True, which="both", axis="y", linestyle="--", alpha=0.7)
plt.ylim(25, 39)
plt.xlim(10**0.3, 10**2.2)

plt.subplot(1, 2, 2)
plt.errorbar(x, ssim_sc_mean, yerr=ssim_sc_std, fmt="-o", color="tab:blue", markersize=6, capsize=4, label="From scratch")
plt.errorbar(x, ssim_unet_mean, yerr=ssim_unet_std, fmt="-s", color="tab:green", markersize=6, capsize=4, label="UNet")
plt.errorbar(x, ssim_fm_mean, yerr=ssim_fm_std, fmt="-^", color="tab:orange", markersize=6, capsize=4, label="FM")
plt.xscale("log")
plt.xlabel("Training Volumes (log scale)")
plt.ylabel("SSIM")
plt.title("SSIM")
plt.legend()
plt.grid(True, which="both", axis="y", linestyle="--", alpha=0.7)
plt.ylim(0.5, 1.0)
plt.xlim(10**0.3, 10**2.2)

plt.tight_layout()
plt.show()
```

```python
psnr_fm_mean = [32.10, 32.69, 32.59]
psnr_fm_std = [5.176, 5.122, 5.200]
ssim_fm_mean = [0.8346, 0.8582, 0.8609]
ssim_fm_std = [0.06275, 0.05668, 0.05469]

x = np.array([300000, 1000000, 3000000])

psnr_base = 30.89
psnr_base_std = 5.757
ssim_base = 0.8246
ssim_base_std = 0.0615

plt.figure(figsize=(10, 4.5))

# --- PSNR subplot ---
plt.subplot(1, 2, 1)
plt.plot(x, psnr_fm_mean, "-^", color="tab:orange", markersize=6, label="Downstream FM (10 volumes)")
plt.fill_between(x, np.array(psnr_fm_mean) - np.array(psnr_fm_std), np.array(psnr_fm_mean) + np.array(psnr_fm_std), color="tab:orange", alpha=0.2)
plt.plot(x, [psnr_base] * len(x), "--", color="black", label="From scratch (10 volumes)")
plt.fill_between(x, [psnr_base - psnr_base_std] * len(x), [psnr_base + psnr_base_std] * len(x), color="gray", alpha=0.2)

plt.xscale("log")
plt.xlabel("Foundation model backbone training volumes (log scale)")
plt.ylabel("PSNR")
plt.title("PSNR")
plt.legend()
plt.grid(True, which="both", axis="y", linestyle="--", alpha=0.7)
plt.ylim(22.5, 42.5)
plt.xlim(10**5.4, 10**6.6)

# --- SSIM subplot ---
plt.subplot(1, 2, 2)
plt.plot(x, ssim_fm_mean, "-^", color="tab:orange", markersize=6, label="Downstream FM (10 volumes)")
plt.fill_between(x, np.array(ssim_fm_mean) - np.array(ssim_fm_std), np.array(ssim_fm_mean) + np.array(ssim_fm_std), color="tab:orange", alpha=0.2)
plt.plot(x, [ssim_base] * len(x), "--", color="black", label="From scratch (10 volumes)")
plt.fill_between(x, [ssim_base - ssim_base_std] * len(x), [ssim_base + ssim_base_std] * len(x), color="gray", alpha=0.2)

plt.xscale("log")
plt.xlabel("Foundation model backbone training volumes (log scale)")
plt.ylabel("SSIM")
plt.title("SSIM")
plt.legend()
plt.grid(True, which="both", axis="y", linestyle="--", alpha=0.7)
plt.ylim(0.7, 1.0)
plt.xlim(10**4.9, 10**6.6)
plt.xlim(10**5.4, 10**6.6)

plt.tight_layout()
plt.show()
```
